### Retrieving the genres from google books api

```python
# Just run this function without making any modifications 

def genre(start,end):
    global l
    l = []
    for i in range(start,end):
        title = books['Title'][i]
        author = books['Author'][i]
        title = str(title)
        author = str(author)
        parms = {"q":"INTITLE:" + "'" + title + "'" + "+" + "INAUTHOR:" + "'" + author + "'", 'key':googleapikey}
        r = requests.get(url="https://www.googleapis.com/books/v1/volumes", params=parms)
        data  = json.loads(r.text)
        if(len(data) > 2):
            for j in range(0, len(data['items'])):
                try:
                    cat = (i,data['items'][j]["volumeInfo"]["categories"][0])
                    break
                except:
                    j+=1
        else:
            cat = (i,'Unknown')
            
        l.append(cat)
        print(cat)
```

```python
labels = ['Index','Genre']
df = pd.DataFrame.from_records(l,columns = labels)
```
